```python
import random
import copy

original = "/content/drive/MyDrive/ForWin/processed/training"
PATH = "/content/drive/MyDrive/Data/processed/"

folder = os.listdir(original)

for path in folder:
  inner_path = os.path.join(original,path)
  
  file_list = os.listdir(inner_path)

  show_list = random.sample(file_list,900)
  l = copy.deepcopy(show_list)

  for index in range(len(l)):
    file_name = l[index]
    source = os.path.join(inner_path,file_name)

    if index < 800:
      
      destination = os.path.join(PATH,"training",path,file_name)
      os.rename(source, destination)
      
    else:

      destination = os.path.join(PATH,"validation",path,file_name)
      os.rename(source, destination)
```
