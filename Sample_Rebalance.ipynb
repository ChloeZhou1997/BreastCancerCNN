```python
import random
import copy

original = "/content/drive/MyDrive/Data/processed/training"
PATH = "/content/drive/MyDrive/Data/processed/"

folder = os.listdir(original)

print(folder)

for path in folder:
  inner_path = os.path.join(original,path)
  file_list = os.listdir(inner_path)
  # print(file_list)

  l = copy.deepcopy(file_list)
  random.shuffle(l)

  length = len(file_list)
  print(length)

  num = 0

  for file_name in l:

    source = os.path.join(inner_path,file_name)
    print(f"The current source path is:{source}")

    
    destination = os.path.join(PATH,"validation",path,file_name)
    print(f"The current destination path is:{destination}")
    os.rename(source, destination)
    print("Path Changed")

    num+=1
    print(f"We just changed path of photo with index:{num} and with file name:{file_name}")
    
    if num > length * 0.2:
      break
```
